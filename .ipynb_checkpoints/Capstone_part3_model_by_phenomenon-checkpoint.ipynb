```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.grid_search import GridSearchCV
from sklearn.pipeline import Pipeline
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import metrics, regularizers
from keras.utils import to_categorical


%matplotlib inline
%config InlineBackend.figure_format = 'retina'
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('2000-2018_disasters_min50k_for_modeling.csv')
df.drop(columns = ['BEGIN_DATE_TIME','END_DATE_TIME','BEGIN_TIME','END_TIME'], inplace = True)
df.head()
```

```python
df['EVENT_TYPE'].value_counts()
```

```python
features = ['BEGIN_RANGE', 'END_RANGE', 'MAGNITUDE', 'BEGIN_LAT','BEGIN_LON', 'END_LAT', 'END_LON',
            'BEGIN_MONTH', 'END_MONTH', 'TOTAL_DURATION_HR', 'DISTANCE_COVERED(KM)', 'BEGIN_HOUR']
len(features)
```

# Hail events

```python
hail = df[df['EVENT_TYPE'] == 'Hail']
hail.shape
```

```python
hail['MAGNITUDE'].min()
```

```python
hail['MAGNITUDE'].max()
```

```python
hail['DAMAGE_PROPERTY'].min()
```

```python
plt.scatter(hail['MAGNITUDE'], hail['DAMAGE_PROPERTY'])
```

```python
plt.scatter(hail['TOTAL_DURATION_HR'], hail['DAMAGE_PROPERTY'])
```

```python
hail = hail[hail['DAMAGE_PROPERTY']<100000000]
plt.scatter(hail['TOTAL_DURATION_HR'], hail['DAMAGE_PROPERTY'])
plt.xlim(-0.2,2)
```

```python
plt.scatter(hail['MAGNITUDE'], hail['DAMAGE_PROPERTY'])
```

```python
X = hail[features]
y = hail['DAMAGE_PROPERTY']

X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
model = Sequential()

# input layer 
model.add(Dense(12, activation = 'relu', input_dim = X_train.shape[1]))

# adding a hidden layer 
model.add(Dense(50, activation = 'relu'))

model.add(Dropout(0.5))

# adding a hidden layer 
model.add(Dense(80, activation = 'relu'))

model.add(Dropout(0.5))

# output layer
model.add(Dense(1, activation = None))

model.compile(loss = 'mean_absolute_error', optimizer = 'adam')

history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 1000, verbose = 0)
```

```python
plt.figure(figsize=(10,8))
plt.plot(history.history['loss'], label = 'train')
plt.plot(history.history['val_loss'], label = 'test')
plt.legend();
```

```python
X = thun[features]
y = thun['DAMAGE_PROPERTY']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
model = Sequential()

# input layer 
model.add(Dense(12, activation = 'relu', input_dim = X_train.shape[1]))

# adding a hidden layer 
model.add(Dense(50, activation = 'relu'))

model.add(Dropout(0.5))

# adding a hidden layer 
model.add(Dense(80, activation = 'relu'))

model.add(Dropout(0.5))

# output layer
model.add(Dense(1, activation = None))

model.compile(loss = 'mean_absolute_error', optimizer = 'adam')

history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 1000, verbose = 0)
```

```python
plt.figure(figsize=(10,8))
plt.plot(history.history['loss'], label = 'train')
plt.plot(history.history['val_loss'], label = 'test')
plt.legend();
```

```python
y_pred = model.predict(X_test)
y_pred = y_pred.reshape(y_pred.shape[0])
y_pred
```

```python
y_test = np.array(y_test)
```

```python
y_pred.shape
```

```python
def coeff_determination(y_true, y_pred):
    from keras import backend as K
    SS_res =  K.sum(K.square( y_true-y_pred ))
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

```python
coeff_determination(np.array(y_test), y_pred)
```
