```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.grid_search import GridSearchCV
from sklearn.pipeline import Pipeline
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import metrics, regularizers
from keras.utils import to_categorical


%matplotlib inline
%config InlineBackend.figure_format = 'retina'
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('2000-2018_disasters_min50k_for_modeling.csv')
df.drop(columns = ['BEGIN_DATE_TIME','END_DATE_TIME','BEGIN_TIME','END_TIME'], inplace = True)
df.head()
```

```python
df.shape
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df = df[df['TOTAL_DURATION_HR'] > 0]
```

```python
features = ['BEGIN_RANGE', 'END_RANGE', 'MAGNITUDE', 'BEGIN_LAT','BEGIN_LON', 'END_LAT', 'END_LON',
            'BEGIN_MONTH', 'END_MONTH', 'TOTAL_DURATION_HR', 'DISTANCE_COVERED(KM)', 'BEGIN_HOUR']
for col in df.columns:
    if col.startswith('STATE_') or col.startswith('EVENT_TYPE_') or col.startswith('MAGNITUDE_TYPE'):
        features.append(col)
        
features
```

```python
df.shape
```

```python
# X = df.drop(columns = ['INJURIES_DIRECT','INJURIES_INDIRECT','DEATHS_DIRECT','DEATHS_INDIRECT','DAMAGE_PROPERTY','DAMAGE_CROPS'])
X = df[features]
y = df['DAMAGE_PROPERTY']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
len(features)
```

# Predicting `DAMAGE_PROPERTY`

## Random Forest

```python
params = {'n_estimators': [250, 300, 350],
          'max_depth': [30,40,50],
          'max_features': [40,50]}
rf = RandomForestRegressor(n_jobs = -1) 
rfr = GridSearchCV(rf, params, cv=5)
rfr.fit(X_train, y_train)
rfr.best_params_
```

```python
rfr.score(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```

# NN with Keras

```python
model = Sequential()

# input layer 
model.add(Dense(80, activation = 'relu', input_dim = X_train.shape[1]))

# adding a hidden layer 
model.add(Dense(100, activation = 'relu'))

model.add(Dropout(0.5))

# output layer
model.add(Dense(1, activation = None))

model.compile(loss = 'mean_absolute_error', optimizer = 'adam')

history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 10000, verbose = 0)
```

```python
plt.figure(figsize=(10,8))
plt.plot(history.history['loss'], label = 'train')
plt.plot(history.history['val_loss'], label = 'test')
plt.legend();
```
